```python
from google.colab import userdata
my_api = userdata.get('GOOGLE_API_KEY')
```

```python
import google.generativeai as genai
```

```python
genai.configure(api_key=my_api)
```

```python
model = genai.GenerativeModel('gemini-1.5-flash')
```

```python
import requests
from PIL import Image
from io import BytesIO
from IPython.display import display

url = 'https://images.pexels.com/photos/1108099/pexels-photo-1108099.jpeg?auto=compress&cs=tinysrgb&dpr=1&w=500'
```

```python
response = requests.get(url)
```

```python
img = Image.open(BytesIO(response.content))
display(img)
```

```python
prompt = """describe picture in detail"""
```

```python
my_response = model.generate_content([prompt,img])
print(my_response.text)
```
